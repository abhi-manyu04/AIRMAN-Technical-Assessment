```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

### Generating fake data for sorties

```python
def generate_flight_training_data(num_rows, output_filename="../data/sorties.csv"):
    
    random.seed(789)
    
    cadets = [f"C{str(i).zfill(3)}" for i in range(1, 21)]          
    instructors = [f"I{str(i).zfill(3)}" for i in range(1, 11)]      
    aircrafts = [f"A{str(i).zfill(3)}" for i in range(1, 13)]        
    bases = ["B01", "B02", "B03"]                                    
    lesson_types = ["Navigation", "Circuit", "General Handling", "Instrument Flying", "Night Flying", "Formation"]
    cancel_reasons = ["Weather", "Aircraft Defect", "Instructor Unavailable", "Cadet Medical", "Airspace Restricted"]
    
    
    start_date = datetime(2026, 5, 1, 6, 0)
    
    data = []
    
    for i in range(1, num_rows + 1):
        sortie_id = f"S{str(i).zfill(3)}"
        cadet_id = random.choice(cadets)
        instructor_id = random.choice(instructors)
        aircraft_id = random.choice(aircrafts)
        base_id = random.choice(bases)
        lesson_type = random.choice(lesson_types)
        
        
        days_offset = random.randint(0, 20)
        hour_offset = random.randint(6, 17) 
        minute_offset = random.choice([0, 15, 30, 45])
        
        sched_start = start_date + timedelta(days=days_offset, hours=hour_offset - start_date.hour, minutes=minute_offset)
        
        duration_minutes = random.choice([60, 90, 120])
        sched_end = sched_start + timedelta(minutes=duration_minutes)
        
        
        status_roll = random.random()
        
        if status_roll < 0.15:
            status = "cancelled"
            delay_minutes = 0
            cancel_reason = random.choice(cancel_reasons)
            actual_start = ""
            actual_end = ""
        else:
            status = "completed"
            cancel_reason = ""
            
            
            delay_roll = random.random()
            if delay_roll < 0.20:
                delay_minutes = random.randint(5, 25)
            elif delay_roll < 0.30:
                delay_minutes = random.randint(26, 60)
            else:
                delay_minutes = 0
                
            act_start_dt = sched_start + timedelta(minutes=delay_minutes)
            
            act_duration = duration_minutes + random.randint(-5, 5)
            act_end_dt = act_start_dt + timedelta(minutes=act_duration)
            
            actual_start = act_start_dt.strftime("%Y-%m-%d %H:%M")
            actual_end = act_end_dt.strftime("%Y-%m-%d %H:%M")
            
        data.append({
            "sortie_id": sortie_id,
            "cadet_id": cadet_id,
            "instructor_id": instructor_id,
            "aircraft_id": aircraft_id,
            "base_id": base_id,
            "scheduled_start": sched_start.strftime("%Y-%m-%d %H:%M"),
            "scheduled_end": sched_end.strftime("%Y-%m-%d %H:%M"),
            "actual_start": actual_start,
            "actual_end": actual_end,
            "status": status,
            "delay_minutes": delay_minutes,
            "cancel_reason": cancel_reason,
            "lesson_type": lesson_type
        })
        
    df = pd.DataFrame(data)
    df.to_csv(output_filename, index=False)
    print(f"Successfully generated and saved {num_rows} records to '{output_filename}'.")


generate_flight_training_data(num_rows=100)
```

### Generating fake data for aircraft

```python
def generate_aircraft_data(num_rows, output_filename="../data/aircraft.csv"):
    
    random.seed(749)
    
    models = ["C172", "PA28", "DA40", "C152", "DA42"]
    bases = ["B01", "B02", "B03"]
    
    used_registrations = set()
    data = []
    
    for i in range(1, num_rows + 1):
        aircraft_id = f"A{str(i).zfill(3)}"
        
    
        while True:
            letters = "".join(random.choices("ABCDEFGHIJKLMNOPQRSTUVWXYZ", k=3))
            registration = f"VT-{letters}"
            if registration not in used_registrations:
                used_registrations.add(registration)
                break
                
    
        ac_type = random.choice(models) if random.random() > 0.03 else ""
        base_id = random.choice(bases) if random.random() > 0.03 else ""
        
    
        if random.random() > 0.05: 
            total_available_hours = random.randint(100, 200)
            defect_count = random.randint(0, 12)
            
        
            if defect_count == 0:
                maintenance_downtime_hours = random.choice([0, 0, 12]) 
            else:
                maintenance_downtime_hours = defect_count * random.randint(4, 8)
        else:
        
            total_available_hours = ""
            maintenance_downtime_hours = ""
            defect_count = ""
            
        data.append({
            "aircraft_id": aircraft_id,
            "registration": registration,
            "type": ac_type,
            "base_id": base_id,
            "total_available_hours": total_available_hours,
            "maintenance_downtime_hours": maintenance_downtime_hours,
            "defect_count": defect_count
        })
        

    df = pd.DataFrame(data)
    df.to_csv(output_filename, index=False)
    print(f"Aircraft data successfully saved to: {output_filename}")

generate_aircraft_data(num_rows=10)
```

### Generate fake data for cadets

```python
def generate_cadet_data(num_rows=20):    
    random.seed(749)
    
    courses = ["PPL", "CPL"]
    bases = ["B01", "B02", "B03"]
    
    base_enrollment_start = datetime(2025, 9, 1)
    data = []
    
    for i in range(1, num_rows + 1):
        cadet_id = f"C{str(i).zfill(3)}"
        name = f"Name{i}" 
        
        course = random.choice(courses) if random.random() > 0.02 else ""
        if course == "PPL":
            total_required_hours = 45
        elif course == "CPL":
            total_required_hours = 200
        else:
            total_required_hours = ""
            
        home_base = random.choice(bases) if random.random() > 0.02 else ""
        
        days_to_add = random.randint(0, 180)
        enroll_date_dt = base_enrollment_start + timedelta(days=days_to_add)
        enrollment_date = enroll_date_dt.strftime("%Y-%m-%d") if random.random() > 0.02 else ""
        
        if total_required_hours != "":
        
            progress_factor = random.uniform(0.0, 1.05)
            total_flown_hours = round(total_required_hours * progress_factor, 1)
        else:
            total_flown_hours = round(random.uniform(5, 40), 1)
            
    
        if random.random() < 0.04:
            total_flown_hours = ""
        if random.random() < 0.03:
            total_required_hours = ""
            
        data.append({
            "cadet_id": cadet_id,
            "name": name,
            "course": course,
            "home_base": home_base,
            "total_required_hours": total_required_hours,
            "total_flown_hours": total_flown_hours,
            "enrollment_date": enrollment_date
        })
        
    df = pd.DataFrame(data)
    df.to_csv("../data/cadets.csv", index=False)
    print("Cadet data successfully saved to: ../data/cadets.csv")

generate_cadet_data(num_rows=20)
```

### Generating fake data for instructors

```python
def generate_instructor_data(num_rows=10):    
    random.seed(749)
    

    aircraft_models = ["C172", "PA28", "DA40", "C152", "DA42"]
    bases = ["B01", "B02", "B03"]
    
    data = []
    
    for i in range(1, num_rows + 1):
        instructor_id = f"I{str(i).zfill(3)}"
        name = f"Name{i}" 
        
        base_id = random.choice(bases) if random.random() > 0.02 else ""
        
    
        if random.random() > 0.03:
            num_quals = random.choice([1, 1, 1, 2])
            quals = random.sample(aircraft_models, k=num_quals)
            aircraft_qualified = ", ".join(quals)
        else:
            aircraft_qualified = ""
            
    
        if random.random() > 0.05:
            total_duty_hours = random.randint(80, 180)
        
            flight_ratio = random.uniform(0.50, 0.75)
            total_flight_hours = int(total_duty_hours * flight_ratio)
        else:
        
            total_duty_hours = ""
            total_flight_hours = ""
            
        data.append({
            "instructor_id": instructor_id,
            "name": name,
            "base_id": base_id,
            "aircraft_qualified": aircraft_qualified,
            "total_duty_hours": total_duty_hours,
            "total_flight_hours": total_flight_hours
        })
        

    df = pd.DataFrame(data)
    df.to_csv("../data/instructors.csv", index=False)
    print(f"Instructor data successfully saved to: ../data/instructors.csv")

generate_instructor_data(num_rows=10)
```

### Generate togastudy data

```python
def generate_cadet_progress_data(num_rows=50):
    random.seed(749)
    
    subject_map = {
        "Meteorology": 25,
        "Navigation": 30,
        "Air Regulations": 28,
        "Aircraft Technical": 35,
        "Human Performance": 20
    }
    
    cadets = [f"C{str(i).zfill(3)}" for i in range(1, 21)]
    

    reference_date = datetime(2026, 5, 15)
    
    data = []
    generated_pairs = set() 
    
    while len(data) < num_rows:
        cadet_id = random.choice(cadets)
        subject = random.choice(list(subject_map.keys()))
        
    
        if (cadet_id, subject) in generated_pairs:
            continue
        generated_pairs.add((cadet_id, subject))
        
        total_chapters = subject_map[subject]
        
    
        if random.random() > 0.04: 
            chapters_completed = random.randint(0, total_chapters)
            
            if chapters_completed == 0:
                avg_quiz_score = "" 
                practice_tests_attempted = 0
            else:
                avg_quiz_score = random.randint(40, 99)
            
                practice_tests_attempted = max(0, int(chapters_completed / random.randint(4, 8)))
                
        
            days_ago = random.randint(0, 20)
            last_active_date = (reference_date - timedelta(days=days_ago)).strftime("%Y-%m-%d")
        else:
        
            chapters_completed = ""
            avg_quiz_score = ""
            practice_tests_attempted = ""
            last_active_date = ""
            
        data.append({
            "cadet_id": cadet_id,
            "subject": subject,
            "chapters_completed": chapters_completed,
            "total_chapters": total_chapters,
            "avg_quiz_score": avg_quiz_score,
            "last_active_date": last_active_date,
            "practice_tests_attempted": practice_tests_attempted
        })
        
    df = pd.DataFrame(data)
    df = df.sort_values(by=["cadet_id", "subject"]).reset_index(drop=True)
    
    df.to_csv("../data/toga_study.csv", index=False)
    print("Academic progress data successfully saved to: ../data/toga_study.csv")

generate_cadet_progress_data(num_rows=50)
```

### Genrate payment data

```python
def generate_cadet_finance_data(num_rows=20):
    random.seed(749)
    
    cadets = [f"C{str(i).zfill(3)}" for i in range(1, 21)]

    reference_date = datetime(2026, 5, 12)
    data = []

    for cadet_id in cadets[:num_rows]:
    
        invoiced_amount = random.choice([180000, 220000, 250000, 350000, 420000, 480000])
    
        behavior_roll = random.random()
        
        if behavior_roll < 0.15:
        
            paid_amount = invoiced_amount
            outstanding_amount = 0
            
            days_ago = random.randint(5, 45)
            last_payment_date = (reference_date - timedelta(days=days_ago)).strftime("%Y-%m-%d")
            
        elif behavior_roll < 0.90:
        
            raw_paid = random.randint(invoiced_amount // 4, invoiced_amount - 20000)
            paid_amount = (raw_paid // 10000) * 10000
            
        
            outstanding_amount = invoiced_amount - paid_amount
            
            days_ago = random.randint(1, 30)
            last_payment_date = (reference_date - timedelta(days=days_ago)).strftime("%Y-%m-%d")
            
        else:
        
            paid_amount = 0
            outstanding_amount = invoiced_amount
            last_payment_date = ""
            
        if paid_amount > 0 and random.random() < 0.03:
            last_payment_date = ""
            
        data.append({
            "cadet_id": cadet_id,
            "invoiced_amount": invoiced_amount,
            "paid_amount": paid_amount,
            "outstanding_amount": outstanding_amount,
            "last_payment_date": last_payment_date
        })
        

    df = pd.DataFrame(data)
    df.to_csv("../data/payments.csv", index=False)
    print("Financial ledger successfully saved to: ../data/payments.csv")

generate_cadet_finance_data()
```
